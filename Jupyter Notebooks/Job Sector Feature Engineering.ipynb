```python
# Find the most common terms in job titles

def combine_job_terms(df_row):
    job_title_terms = df_row['Job Title'].split()
    company_name_terms = df_row['Company Name'].split()
    return job_title_terms + company_name_terms

# Create a new column of job terms to search for
df['Job Terms'] = df.apply(combine_job_terms, axis=1)
print(df['Job Terms'])
```

```python
# Count the most frequent terms. Used for building job sector dictionary
from itertools import combinations
from collections import Counter

count = Counter()

for row in df['Job Terms']:
    count.update(Counter(row))
    
for key, value, in count.most_common():
    print(key, value)
```

```python
# Set the Job Sector column
df["Job Sector"] = ''
df["Job Sector"] = df["Job Sector"].apply(list)
```

```python
def find_and_input_job_sector(job_sector_dict, df):
    for keyword, job_sector in job_sector_dict.items():
        # Find job postings based on job title or company name containing a keyword 
        mask = (df["Job Title"].str.contains(keyword) | df["Company Name"].str.contains(keyword))

        # Fill in the job sector based on the criteria above
        
        for cell in df.loc[mask, 'Job Sector']:
            if job_sector not in cell:
                cell.append(job_sector)
        #print(df.loc[mask, 'Job Sector'])
        #df.loc[mask, 'Job Sector'] = df.loc[mask, 'Job Sector'].append(list(job_sector))
    return df
```

```python
df = find_and_input_job_sector(job_sector_dict, df)
```
