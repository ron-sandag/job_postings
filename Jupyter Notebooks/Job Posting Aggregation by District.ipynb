```python
import pandas as pd
from SD_District_Mapping import find_geolocation_ZIP_Code
```

```python
def find_ZIP_Code(row):
    if pd.isnull(row['ZIP Code']):
        return find_geolocation_ZIP_Code(row['Location'])
    else:
        return row['ZIP Code']
```

```python
# Import our .csv file
df = pd.read_csv('../Data/Job_Posting_Data_With_Job_Sectors.csv')

# Turn the 'Job Sector' column into a list
df['Job Sector'] = df['Job Sector'].apply(eval)

# Create a column of ZIP codes
df['ZIP Code'] = df['Location'].str.extract(r'(\d{5})')

# Use 'Location' column to find additional ZIP codes
df['ZIP Code'] = df.apply(find_ZIP_Code, axis=1)

#temp = df.apply(find_ZIP_Code, axis=1)
#print(temp)
print(df)
```

```python
# Create District attribute
import os
from SD_District_Mapping import find_SD_district

zip_code_aggregates_df['District'] = zip_code_aggregates_df.index.to_series().apply(find_SD_district)
print(zip_code_aggregates_df)
zip_code_aggregates_df.to_csv('../Aggregates/Zip_Code_Aggregates.csv')
```

```python
# Create District aggregate DataFrame
district_aggregates_df = zip_code_aggregates_df.groupby(['District']).sum()

# The Median Salary column is incorrect, so we drop it and reaggregate
district_aggregates_df.pop('Median Salary')
district_aggregates_df['Median Salary'] = zip_code_aggregates_df.groupby(['District'])['Median Salary'].mean()
print(district_aggregates_df)
```

```python
# Add a location column for geolocating in PowerBI
district_aggregates_df['Location ZIP Code'] = district_aggregates_df.index.to_series().apply(find_geolocation_ZIP_Code)
print(district_aggregates_df)
```

```python
# Remove districs that have less than 5 job postings
district_aggregates_df = district_aggregates_df[district_aggregates_df["Job Openings"] > 4]
print(district_aggregates_df)
```
